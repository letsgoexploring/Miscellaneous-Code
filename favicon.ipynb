```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
plt.style.use('ggplot')
```

```python
x = np.arange(-10,10,0.001)
y = np.sin(x)/x+x/15

fig = plt.figure(figsize=(5,5))
ax = fig.add_subplot(1,1,1)
ax.fill_between(x, y, 0*y-1 , facecolor='blue', alpha=0.3)
ax.set_xticklabels([])
ax.set_yticklabels([])
ax.plot(x,y,lw=5)
# ax.grid()

plt.savefig('favicon.png',dpi=120,bbox_inches='tight')
```
